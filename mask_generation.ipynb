```python
import os
import cv2
import numpy as np
from tensorflow.keras.utils import to_categorical

IMG_SIZE = 512
NUM_CLASSES = 3

def load_data(image_dir, mask_dir):
    images = []
    masks = []

    for filename in os.listdir(image_dir):
        img_path = os.path.join(image_dir, filename)
        mask_path = os.path.join(mask_dir, filename.replace('.jpg', '.png'))

        if not os.path.exists(mask_path):
            continue

        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = img / 255.0

        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        mask = cv2.resize(mask, (IMG_SIZE, IMG_SIZE))
        mask = to_categorical(mask, NUM_CLASSES)

        images.append(img)
        masks.append(mask)

    return np.array(images), np.array(masks)
```

```python
import os
import cv2
import numpy as np
from pycocotools.coco import COCO

IMG_DIR  = "/content/drive/MyDrive/Colab Notebooks/train/images"
ANN_FILE = "/content/drive/MyDrive/Colab Notebooks/train/_annotations.coco.json"
MASK_DIR = "/content/drive/MyDrive/Colab Notebooks/train/masks"

os.makedirs(MASK_DIR, exist_ok=True)

coco = COCO(ANN_FILE)

CLASS_MAP = {
    "ruins": 1,
    "vegetation": 2
}

for img_id in coco.getImgIds():
    img = coco.loadImgs(img_id)[0]
    mask = np.zeros((img["height"], img["width"]), dtype=np.uint8)

    for ann in coco.loadAnns(coco.getAnnIds(imgIds=img_id)):
        cat = coco.loadCats(ann["category_id"])[0]["name"]
        if cat in CLASS_MAP:
            m = coco.annToMask(ann)
            mask[m == 1] = CLASS_MAP[cat]

    mask_name = os.path.splitext(img["file_name"])[0] + ".png"
    cv2.imwrite(os.path.join(MASK_DIR, mask_name), mask)

print("✅ Masks generated")
```

```python
import os
import cv2
import numpy as np
from tensorflow.keras.utils import to_categorical

IMG_PATH  = "/content/drive/MyDrive/Colab Notebooks/train/images"
MASK_PATH = "/content/drive/MyDrive/Colab Notebooks/train/masks"

IMG_SIZE = 512
NUM_CLASSES = 3

def load_data(img_dir, mask_dir):
    X, Y = [], []

    for img_name in os.listdir(img_dir):
        img_file = os.path.join(img_dir, img_name)
        mask_file = os.path.join(mask_dir, os.path.splitext(img_name)[0] + ".png")

        if not os.path.exists(mask_file):
            continue

        img = cv2.imread(img_file)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE)) / 255.0

        mask = cv2.imread(mask_file, cv2.IMREAD_GRAYSCALE)
        mask = cv2.resize(mask, (IMG_SIZE, IMG_SIZE))
        mask = to_categorical(mask, NUM_CLASSES)

        X.append(img)
        Y.append(mask)

    return np.array(X), np.array(Y)

X_train, y_train = load_data(IMG_PATH, MASK_PATH)

print("X shape:", X_train.shape)
print("Y shape:", y_train.shape)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def unet(input_shape=(512, 512, 3), num_classes=3):
    inputs = layers.Input(input_shape)

    # Encoder
    c1 = layers.Conv2D(64, 3, activation='relu', padding='same')(inputs)
    c1 = layers.Conv2D(64, 3, activation='relu', padding='same')(c1)
    p1 = layers.MaxPooling2D()(c1)

    c2 = layers.Conv2D(128, 3, activation='relu', padding='same')(p1)
    c2 = layers.Conv2D(128, 3, activation='relu', padding='same')(c2)
    p2 = layers.MaxPooling2D()(c2)

    # Bottleneck
    c3 = layers.Conv2D(256, 3, activation='relu', padding='same')(p2)
    c3 = layers.Conv2D(256, 3, activation='relu', padding='same')(c3)

    # Decoder
    u4 = layers.UpSampling2D()(c3)
    u4 = layers.concatenate([u4, c2])
    c4 = layers.Conv2D(128, 3, activation='relu', padding='same')(u4)

    u5 = layers.UpSampling2D()(c4)
    u5 = layers.concatenate([u5, c1])
```

```python
def unet(input_shape=(512, 512, 3), num_classes=3):
    inputs = keras.Input(shape=input_shape)

    c1 = layers.Conv2D(64, 3, activation="relu", padding="same")(inputs)
    c1 = layers.Conv2D(64, 3, activation="relu", padding="same")(c1)
    p1 = layers.MaxPooling2D()(c1)

    c2 = layers.Conv2D(128, 3, activation="relu", padding="same")(p1)
    c2 = layers.Conv2D(128, 3, activation="relu", padding="same")(c2)
    p2 = layers.MaxPooling2D()(c2)

    c3 = layers.Conv2D(256, 3, activation="relu", padding="same")(p2)
    c3 = layers.Conv2D(256, 3, activation="relu", padding="same")(c3)

    u4 = layers.UpSampling2D()(c3)
    u4 = layers.Concatenate()([u4, c2])
    c4 = layers.Conv2D(128, 3, activation="relu", padding="same")(u4)

    u5 = layers.UpSampling2D()(c4)
    u5 = layers.Concatenate()([u5, c1])
    c5 = layers.Conv2D(64, 3, activation="relu", padding="same")(u5)

    outputs = layers.Conv2D(num_classes, 1, activation="softmax")(c5)

    return keras.Model(inputs, outputs)
```

```python
class SegmentationDataset(Dataset):
    def __init__(self, image_dir, mask_dir, img_size=256):
        self.image_dir = image_dir
        self.mask_dir = mask_dir
        self.images = sorted(os.listdir(image_dir))
        self.masks  = sorted(os.listdir(mask_dir))
        self.img_size = img_size

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path  = os.path.join(self.image_dir, self.images[idx])
        mask_path = os.path.join(self.mask_dir, self.masks[idx])

        image = cv2.imread(img_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = cv2.resize(image, (self.img_size, self.img_size))
        image = image / 255.0

        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        mask = cv2.resize(mask, (self.img_size, self.img_size))

        # IMPORTANT: mask must be Long (class labels)
        return (
            torch.tensor(image).permute(2,0,1).float(),
            torch.tensor(mask).long()
        )
```

```python
class DoubleConv(nn.Module):
    def __init__(self, in_c, out_c):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_c, out_c, 3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_c, out_c, 3, padding=1),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.conv(x)


class UNet(nn.Module):
    def __init__(self, num_classes=3):
        super().__init__()

        self.d1 = DoubleConv(3, 64)
        self.d2 = DoubleConv(64, 128)
        self.d3 = DoubleConv(128, 256)

        self.pool = nn.MaxPool2d(2)

        self.b = DoubleConv(256, 512)

        self.u1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.c1 = DoubleConv(512, 256)

        self.u2 = nn.ConvTranspose2d(256, 128, 2, 2)
        self.c2 = DoubleConv(256, 128)

        self.u3 = nn.ConvTranspose2d(128, 64, 2, 2)
        self.c3 = DoubleConv(128, 64)

        # ✅ MULTI-CLASS OUTPUT
        self.out = nn.Conv2d(64, num_classes, 1)

    def forward(self, x):
        c1 = self.d1(x)
        c2 = self.d2(self.pool(c1))
        c3 = self.d3(self.pool(c2))

        b = self.b(self.pool(c3))

        u1 = self.u1(b)
        u1 = self.c1(torch.cat([u1, c3], dim=1))

        u2 = self.u2(u1)
        u2 = self.c2(torch.cat([u2, c2], dim=1))

        u3 = self.u3(u2)
        u3 = self.c3(torch.cat([u3, c1], dim=1))

        return self.out(u3)   # ❌ NO sigmoid / softmax
```

```python
total_epochs = 25

for epoch in range(start_epoch, total_epochs):
    model.train()
    epoch_loss = 0

    for images, masks in loader:
        images = images.to(device)
        masks = masks.to(device)

        outputs = model(images)
        loss = criterion(outputs, masks)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

    avg_loss = epoch_loss / len(loader)
    print(f"Epoch {epoch+1}/{total_epochs}, Loss: {avg_loss:.4f}")

    save_path = f"{PROJECT_DIR}/unet_epoch_{epoch+1}.pth"
    torch.save({
        "epoch": epoch + 1,
        "model_state_dict": model.state_dict(),
        "optimizer_state_dict": optimizer.state_dict(),
        "loss": avg_loss
    }, save_path)

    print(f"✅ Model saved to Drive: {save_path}")
```

```python
def dice_score(pred, target):
    intersection = (pred == target).sum().float()
    return (2 * intersection) / (pred.numel() + target.numel() + 1e-6)

def iou_score(pred, target):
    intersection = ((pred == target) & (target > 0)).sum().float()
    union = ((pred > 0) | (target > 0)).sum().float()
    return intersection / (union + 1e-6)
```

```python
import os
import numpy as np
from PIL import Image
import torch
import matplotlib.pyplot as plt

# Paths
IMAGE_DIR = "/content/drive/MyDrive/Colab Notebooks/UNet_Project/train/images"
MASK_DIR  = "/content/drive/MyDrive/Colab Notebooks/UNet_Project/train/masks"

RESULTS_DIR = "/content/drive/MyDrive/unet_project/evaluation_results"
os.makedirs(RESULTS_DIR, exist_ok=True)

# U-Net model in eval mode
model.eval()

dice_scores = []
iou_scores = []

for img_name in os.listdir(IMAGE_DIR):

    img_path  = os.path.join(IMAGE_DIR, img_name)
    mask_path = os.path.join(MASK_DIR, img_name)  # assuming filenames match

    # Load image and mask
    image = Image.open(img_path).convert("RGB")
    mask  = Image.open(mask_path)

    # Preprocess image
    image_tensor = transform(image).unsqueeze(0).to(device)
    mask_tensor  = torch.tensor(np.array(mask)).to(device)

    # Prediction
    with torch.no_grad():
        output = model(image_tensor)
        pred = torch.argmax(output, dim=1)

    # Dice score
    intersection = ((pred == mask_tensor) & (mask_tensor > 0)).sum().float()
    union = (pred > 0).sum().float() + (mask_tensor > 0).sum().float()
    dice = (2 * intersection) / (union + 1e-8)
    dice_scores.append(dice.item())

    # IoU score
    iou = intersection / (union - intersection + 1e-8)
    iou_scores.append(iou.item())

    # Save predicted mask
    save_path = os.path.join(RESULTS_DIR, img_name.replace(".jpg", "_pred.png"))
    plt.imsave(save_path, pred[0].cpu().numpy(), cmap="gray")

# Save evaluation scores
with open(os.path.join(RESULTS_DIR, "evaluation_scores.txt"), "w") as f:
    f.write(f"Average Dice Score: {sum(dice_scores)/len(dice_scores)}\n")
    f.write(f"Average IoU Score: {sum(iou_scores)/len(iou_scores)}\n")

print("✅ Evaluation completed. Results saved in Google Drive.")
```

```python
import os
import numpy as np
from PIL import Image
import torch
import matplotlib.pyplot as plt

# Paths
IMAGE_DIR = "/content/drive/MyDrive/Colab Notebooks/UNet_Project/train/images"
MASK_DIR  = "/content/drive/MyDrive/Colab Notebooks/UNet_Project/train/masks"
RESULTS_DIR = "/content/drive/MyDrive/unet_project/evaluation_results"
os.makedirs(RESULTS_DIR, exist_ok=True)

# U-Net model in eval mode
model.eval()

dice_scores = []
iou_scores = []

# Get all mask filenames
mask_files = os.listdir(MASK_DIR)
mask_dict = {os.path.splitext(f)[0]: f for f in mask_files}  # key = filename without extension

for img_name in os.listdir(IMAGE_DIR):

    img_path = os.path.join(IMAGE_DIR, img_name)
    img_base = os.path.splitext(img_name)[0]

    # Find corresponding mask
    if img_base in mask_dict:
        mask_path = os.path.join(MASK_DIR, mask_dict[img_base])
    else:
        print(f"⚠️ Mask not found for {img_name}, skipping...")
        continue

    # Load image and mask
    image = Image.open(img_path).convert("RGB")
    mask  = Image.open(mask_path)

    # Preprocess image
    image_tensor = transform(image).unsqueeze(0).to(device)
    mask_tensor  = torch.tensor(np.array(mask)).to(device)

    # Prediction
    with torch.no_grad():
        output = model(image_tensor)
        pred = torch.argmax(output, dim=1)

    # Dice score
    intersection = ((pred == mask_tensor) & (mask_tensor > 0)).sum().float()
    union = (pred > 0).sum().float() + (mask_tensor > 0).sum().float()
    dice = (2 * intersection) / (union + 1e-8)
    dice_scores.append(dice.item())

    # IoU score
    iou = intersection / (union - intersection + 1e-8)
    iou_scores.append(iou.item())

    # Save predicted mask
    save_path = os.path.join(RESULTS_DIR, img_base + "_pred.png")
    plt.imsave(save_path, pred[0].cpu().numpy(), cmap="gray")

# Save evaluation scores
with open(os.path.join(RESULTS_DIR, "evaluation_scores.txt"), "w") as f:
    f.write(f"Average Dice Score: {sum(dice_scores)/len(dice_scores)}\n")
    f.write(f"Average IoU Score: {sum(iou_scores)/len(iou_scores)}\n")

print("✅ Evaluation completed. All results saved in Google Drive.")
```
